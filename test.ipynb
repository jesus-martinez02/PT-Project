```python
df_annoying = df.loc[
        (df['trip_id'].isin(list_trip_id))
        & ~(
          (df['stop_name'] == 'Södersjukhuset')
            | ((df['stop_name'] == 'Rosenlund') & (df['stop_headsign'] == 'Sickla udde via Södersjukhuset'))
            | ((df['stop_name'] == 'Södra station') & (df['stop_headsign'] == 'Hornsberg via Södersjukhuset'))
            )
         ]

df_annoying['stop_sequence_real'] = df_annoying['stop_sequence_real'] .apply(lambda x: x-1 if x not in ['Sickla udde via Södersjukhuset','Hornsberg via Södersjukhuset'] else x)

df_annoying['stop_headsign'] = df_annoying['stop_headsign'].apply(lambda des: 'Sickla udde' if des =='Sickla udde via Södersjukhuset'  else ('Hornsberg' if des == 'Hornsberg via Södersjukhuset' else des))

df_clean = pd.concat([df_annoying,df[~df['trip_id'].isin(list_trip_id)]])
```

```python
ordered_df['distance_between_stops'] = ordered_df.groupby('trip_id')['shape_dist_traveled'].diff()
ordered_df
```

```python
ordered_df['arrival_time_real'] = pd.to_datetime(ordered_df['arrival_time_real'], errors='coerce')
ordered_df['departure_time_real'] = pd.to_datetime(ordered_df['departure_time_real'], errors='coerce')

ordered_df['prev_departure_real'] = ordered_df.groupby('trip_id')['departure_time_real'].shift(1)

# Calculate time between stops (travel time)
ordered_df['travel_time'] = (ordered_df['arrival_time_real'] - ordered_df['prev_departure_real']).dt.total_seconds()

ordered_df['speed'] = (ordered_df['distance_between_stops'] / 1000) / (ordered_df['travel_time'] / 3600)

ordered_df
```
